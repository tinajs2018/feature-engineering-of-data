```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
```

```python
df=pd.read_csv("melb_data.csv")
df.head(4)
```

```python
 len(df['Address'].unique())
```

```python
df.isna().sum()
```

```python
df['BuildingArea'].mean()
df['area']=df['BuildingArea'].fillna(151.96)
```

```python
df['Car'].mean()
df['Cars']=df['Car'].fillna(1.61)
```

```python
df['YearBuilt'].describe()
df['yearbuilt']=df['YearBuilt'].fillna(1999)
```

```python
# df['council']  =df['CouncilArea'].fillna(method='bfill',inplace=True)
```

```python
new=df.fillna(method='bfill',axis=0)
new
```

```python
new.isnull().sum()
```

```python
data= new.drop(['CouncilArea'],axis=1)
data.head(3)
```

```python
data.isnull().sum()
```

```python
plt.figure(figsize=(16,8))
sns.heatmap(data.corr(),annot=True)
```

```python
#finding the data which is correlated
real=data.corr()
real['Price'].sort_values(ascending=False)
```

Most correlated variable to the target varaibles are Rooms 0.496634 Bedroom2 0.475951 Bathroom 0.467038 Cars 0.238637 Car 0.238096 Longtitude 0.203656 Postcode

```python
x=( data[['Rooms','Distance','Postcode','Bedroom2','Landsize','BuildingArea','YearBuilt','area','yearbuilt']])
x.head(1)
```

```python
y=data['Price']
y.head(1)
```

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
bestfeature= SelectKBest(score_func=chi2,k=5)
fit=bestfeature.fit(x,y)
fit
```

```python
dfscore=pd.DataFrame(fit.scores_)
dfcolumns=pd.DataFrame(x.columns)
Featurescore=pd.concat([dfcolumns,dfscore],axis=1)
Featurescore.columns=['specks','score']
Featurescore
```

From the above method  it show distance has the highest relation to the traget varaible 

```python
from sklearn.ensemble import ExtraTreesClassifier
import matplotlib.pyplot as plt
model=ExtraTreesClassifier()
model.fit(x,y)
print(model.feature_importances_)
```
